# Prediction File

```python
# predict.py

import pickle
import pandas as pd
```

```python
# Load the saved model
with open("model.pkl", "rb") as file:
    model = pickle.load(file)

```

## Individual Prediction

```python
# Ask the user for input values
phi = float(input("Enter value for phi: "))
Ha = float(input("Enter value for Ha: "))
Ra = float(input("Enter value for Ra: "))

# Create a DataFrame with the same feature names used during training
X_new = pd.DataFrame({
    'phi': [phi],
    'Ha': [Ha],
    'Ra': [Ra]
})

# Make prediction
prediction = model.predict(X_new)

# Display the result
print(f"Predicted Nu: {prediction[0]}")
```

## For a whole dataset

```python
# Read the CSV file
df = pd.read_csv("input_file.csv")
```

```python
# Check that required columns exist
required_columns = ['phi', 'Ha', 'Ra']

missing_columns = [col for col in required_columns if col not in df.columns]
if missing_columns:
    raise ValueError(f"Missing required columns: {missing_columns}")

# Select only the independent variables
X = df[required_columns]
```

```python
# Predict Nu for all rows
df['Nu_predicted'] = model.predict(X)
```
